### Commandes pour les tableaux

```python
import simplificator as sf
tableau = sf.charger_donnees_trafic_1()
```

* Afficher un tableau :

```python
n = 15 #nombre de lignes à afficher
tableau.afficher(n)
```

* Afficher le nombre de lignes / colonnes :

```python
tableau.afficher_le_nombre_de_lignes()
tableau.afficher_le_nombre_de_colonnes()
```

* Afficher les valeurs d'une colonne :

```python
nom = 'etat_trafic' #nom de la colonne
tableau.afficher_les_valeurs_de_la_colonne(nom)
```

* Afficher le nom des colonnes :

```python
tableau.afficher_le_nom_des_colonnes()
```

* Avoir un élément

```python
l = 0 #numéro de la ligne
c = 'etat_trafic' #nom de la colonne
element = tableau.avoir_element(l, c)
print(element)
```

* Avoir un sous-tableau

```python
lignes = [0, 25, 34] #lignes à garder
colonnes = ['etat_trafic', 'id_troncon'] #colonnes à garder
sous_tableau = tableau.sous_tableau(lignes=lignes, colonnes=colonnes)
sous_tableau.afficher()
```

* Coller un tableau à un autre 

```python
tableau_1 = tableau.sous_tableau(colonnes=['etat_trafic'])
tableau_2 = tableau.sous_tableau(colonnes=['id_troncon'])

tableau_3 = tableau_1.coller(tableau_2)
tableau_3.afficher()
```

* Créer un nouveau tableau

```python
valeurs = [['a', 'A'], ['b', 'B']]
noms_colonnes = ['minuscule', 'majuscule']
tableau_alphabet = sf.nouveau_tableau(valeurs, noms_colonnes)
tableau_alphabet.afficher()
```

### Commandes pour l'apprentissage

* Créer une IA

```python
akinator = sf.SfNaiveBayes()
```

* L'entraîner avec des données `entree` et `sortie`

```python
donnees_entrainement = sf.charger_donnees_trafic_2()
entree = donnees_entrainement.sous_tableau(colonnes=['denomination', 'jour', 'heure'])
sortie = donnees_entrainement.sous_tableau(colonnes=['etat_trafic'])
akinator.entrainer(entree, sortie)
```

* Prédire une sortie avec une nouvelle entrée

```python
nouvelle_entree = sf.nouveau_tableau(lignes=[['Route départementale 34', '15', 'jeudi']], colonnes=['denomination', 'jour', 'heure'])
sortie_predite = akinator.predire(nouvelle_entree)
sortie_predite.afficher()
```
